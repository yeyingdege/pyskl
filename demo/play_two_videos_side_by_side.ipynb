# Play Two Videos Side by Side

Set `video_left` and `video_right` to local notebook-accessible paths or URLs, then run the display cell.

Examples:
- `video_left = "video1.mp4"`
- `video_right = "./clips/video2.webm"`
- `video_left = "https://example.com/video1.mp4"`

For local files, place the videos in the same folder as this notebook or use a relative path.

```python
# Change these to your video paths or URLs
video_left = "/home/yanghong/source/pyskl/output/demo_output_027_fft_seed0_yolo26l.mp4"
video_right = "/home/yanghong/source/pyskl/output/demo_output_027_fft_seed0_yolo26l_d2_ep10.mp4"

# Optional labels shown above each video
label_left = "YOLO26l-pose"
label_right = "YOLO26l-pose finetuned on keypoints dataset v2"
```

```python
from IPython.display import HTML, display
from html import escape

def show_videos_side_by_side(left_src, right_src, left_label="Video 1", right_label="Video 2", width="48%"):
    """Display two videos side by side in a Jupyter notebook.

    The Play Both, Pause Both, and Reset buttons control both videos together.
    Browser support and local file serving behavior may vary by Jupyter environment.
    """
    left_src = escape(left_src, quote=True)
    right_src = escape(right_src, quote=True)
    left_label = escape(left_label)
    right_label = escape(right_label)
    width = escape(width, quote=True)

    html = f"""
    <div style="font-family: Arial, sans-serif;">
      <div style="display: flex; gap: 16px; align-items: flex-start; justify-content: center;">
        <div style="width: {width}; text-align: center;">
          <h3 style="margin-bottom: 8px;">{left_label}</h3>
          <video id="video_left" src="{left_src}" controls style="width: 100%; max-height: 75vh; background: #000;"></video>
        </div>
        <div style="width: {width}; text-align: center;">
          <h3 style="margin-bottom: 8px;">{right_label}</h3>
          <video id="video_right" src="{right_src}" controls style="width: 100%; max-height: 75vh; background: #000;"></video>
        </div>
      </div>

      <div style="margin-top: 14px; text-align: center;">
        <button onclick="playBothVideos()">Play Both</button>
        <button onclick="pauseBothVideos()">Pause Both</button>
        <button onclick="resetBothVideos()">Reset Both</button>
      </div>
    </div>

    <script>
      function getVideos() {{
        return [document.getElementById('video_left'), document.getElementById('video_right')];
      }}

      function playBothVideos() {{
        const [left, right] = getVideos();
        left.play();
        right.play();
      }}

      function pauseBothVideos() {{
        const [left, right] = getVideos();
        left.pause();
        right.pause();
      }}

      function resetBothVideos() {{
        const [left, right] = getVideos();
        left.pause();
        right.pause();
        left.currentTime = 0;
        right.currentTime = 0;
      }}
    </script>
    """
    display(HTML(html))

show_videos_side_by_side(video_left, video_right, label_left, label_right)
```
